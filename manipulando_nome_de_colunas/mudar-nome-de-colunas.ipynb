```python
import pandas as pd
```

```python
df = pd.DataFrame(
    {'Nome  ': ["Arthur", "Júlia", "Paulo"],
     '  aGe  ': [26, 1, 30]}
)
```

# Transformar todas colunas

```python
df.columns = df.columns.str.lower() # upper() para maiúsculo
df.columns = df.columns.str.strip()
```

```python
coluna_renomeada = {
    "age": "idade"
}

df.rename(columns=coluna_renomeada, inplace=True)
```
