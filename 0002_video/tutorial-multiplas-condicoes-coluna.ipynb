# Bibliotecas e módulos

```python
import pandas as pd
import numpy as np
```

# Carregando dataset

```python
data_dict = {
    'col1': [7, 8, 5, 10],
    'col2': [2, 9, 5, -100]
}

df = pd.DataFrame(data_dict)
df
```

# Trabalhando com condicionais

## Duas condições

```python
df['cond1'] = np.where(
    df['col1'] > df['col2'],
    df['col1'] + df['col2'],
    df['col1'] - df['col2']
)
```

```python
df
```

## Múltiplas condições com np.where

```python
multi_cond = np.where(
    df['col1'] == df['col2'],
    df['col1'] * df['col2'],
    np.where(
        df['col1'] > df['col2'],
        df['col1'] + df['col2'],
        df['col1'] - df['col2']
    )
)

df['cond2'] = multi_cond
```

```python
df
```

## Múltiplas condições com np.select

```python
condicoes = [
    df['col2'] < 0,
    df['col1'] == df['col2'],
    df['col1'] > df['col2']
]

resultados = [
    df['col2'] * -1,
    df['col1'] * df['col2'],
    df['col1'] + df['col2']
]

df['col_select'] = np.select(condicoes, resultados, df['col1'] - df['col2'])
df
```
