```python
import pandas as pd
```

```python
date_dict = {
    "data_padrao": ["2021-07-02 19:07", "2020-01-31 15:00", "2013-09-09 08:00"],
    "data_formato_br": ["21/09/1995 10:00", "10/12/2001 08:57", "01/01/2000 00:01"]
}
```

```python
df = pd.DataFrame(date_dict)
```

```python
df
```

# Convertendo texto para data

```python
df.dtypes
```

```python
df["data_padrao"] = pd.to_datetime(df["data_padrao"])
```

```python
df
```

```python
formato = '%d/%m/%Y %H:%M'
df["data_formato_br"] = pd.to_datetime(df["data_formato_br"], format=formato)
```
